# Matplotlib-2

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.arange(10,20)
b=np.arange(25,30)
print(a)
print(b)
```

```python
z=list(zip(a,b))
z
```

```python
x=np.arange(10)
x
```

```python
for i in z:
#     print(i)
    plt.scatter(i[0],i[1])
plt.xlabel("data of a")
plt.ylabel("data of b")
plt.title("A and B")
plt.show()
```

```python
x
```

```python
y=np.arange(10,20)
y
```

```python
plt.scatter(x,y,s=200,c="#00897B",alpha=0.3,marker=">")
plt.xlabel("X Axis-->")
plt.ylabel("Y Axis-->")
plt.title("Matplotlib Demo")
plt.show()
```

```python
plt.plot(x,y,c="red",marker="x",linestyle=":",linewidth=5,alpha=0.5)
```

```python
plt.figure(figsize=(10,9))
plt.plot(x,y,"-.r")
plt.xlim(-5,15)
plt.ylim(-4,20)
```
